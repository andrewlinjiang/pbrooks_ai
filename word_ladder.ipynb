```python
with open("dictall.txt", "r") as fp:
    content = fp.read().splitlines() 
content = sorted(content,key = len)


def order(length):
    word = []
    for i in content:
        if len(i) == length:
            word.append(i)
    return word # returns an array of all "length" letter words in alphabetical order

                

def off_by_one(a,b): 
    count = -1
    for i in range(len(a)):
        if a[i] != b[i]:
            count += 1
    return(not bool(count))


def connect(lis): #returns a dictionary where each key are the words and each value are adjacent words
    node = {word:[] for word in lis}# don't need to create neighbors for all n letter words, make it as you go
    abc = "qwertyuiopasdfghjklzxcvbnm"
    lis = set(lis)
    for i in lis:
        for j in range(len(i)):
            for letter in abc:
                diff_by_one = i[:j] + letter + i[j+1:]
                #print(diff_by_one)
                if diff_by_one in lis and diff_by_one != i:
                    node[i].append(diff_by_one)
    
    return node

def h(a,b): #h of x estimate
    count = 0
    for i in range(len(a)):
        if a[i] != b[i]:
            count += 1
    return count

print(connect(order(4)))
```

```python
import heapq as hq 

class Node:
  
    def __init__(self, data):
        self.word = data
        self.f = 0
        self.h = 0
        self.g = 0
        
        self.previous = None
        
    def value(self):
        return self.data

    
def A(start,end):
    count = 0
    dic = connect(order(len(start)))
    frontier = [] #heap
    explored = set() #set
    
    start_node = Node(start)
    
    start_node.h = h(start, end)
    start_node.g = 0
    start_node.f = start_node.g + start_node.h
    hq.heappush(frontier,(int(start_node.f),count,  start_node))
    
    #print(dic)
    while len(frontier):

        current = hq.heappop(frontier)
        node_current = current[2]
        if node_current.word == end :
            return create_path(node_current)
        else:
            explored.add(node_current.word)
            neighbors = dic[node_current.word]
            for i in neighbors:
                count += 1
                new_g = node_current.g + 1
                node_neighbor = Node(i)
                node_neighbor.g = new_g
                node_neighbor.h = h(i,end)
                node_neighbor.f = new_g + node_neighbor.h
                node_neighbor.previous = node_current
                #print(node_current.word, node_neighbor.previous.word)
                if i not in explored:
                    hq.heappush(frontier,(int(node_neighbor.f),count, node_neighbor))
    return "no path"

def create_path(current):
    path = []
    while current.previous != None:
        #print(current.previous.word)
        path.append(current.word)
        current = current.previous
    path.append(current.word)
    return (len(path), path[::-1])
            
print(A("fold", "weal"))
print(A("string", "abased"))
print(A("string", "abased"))

    
    
```
